# # classify Movie Based on Rating (Excellent, Good and Bad)

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
dataset['Rating_category']=dataset['IMDb_rating'].apply(rating)
```

# # Find unique values from Genre

```python
dataset['genre_1']
```

```python
list1=[]
for value in dataset['genre_1']:
    list1.append (value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
